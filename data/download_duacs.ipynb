Downloads L4 DUACS $u$ and $v$ geostrophic velocity components from CMEMS and store it as a Zarr dataset.

```python
import os

import copernicusmarine as cm
import xarray as xr
```

```python
era5_ds = xr.open_zarr("ERA5-ekman.zarr")
```

```python
era5_ds
```

```python
start_datetime_str = "2023-03-28"
end_datetime_str = "2023-07-11"

min_lat = era5_ds.latitude.min().item() - 1/16
max_lat = era5_ds.latitude.max().item() + 1/16
min_lon = era5_ds.longitude.min().item() - 1/16
max_lon = era5_ds.longitude.max().item() + 1/16

output_directory = "./"
```

```python
dataset_id = "cmems_obs-sl_eur_phy-ssh_my_allsat-l4-duacs-0.0625deg_P1D"
output_filename = f"{dataset_id}_{start_datetime_str}_{end_datetime_str}.zarr"

if not os.path.exists(os.path.join(output_directory, output_filename)):
    cm.subset(
        dataset_id, 
        variables=["ugos", "vgos"],
        start_datetime=start_datetime_str,
        end_datetime=end_datetime_str,
        minimum_latitude=min_lat,
        maximum_latitude=max_lat,
        minimum_longitude=min_lon,
        maximum_longitude=max_lon,
        output_filename=output_filename,
        output_directory=output_directory
    )
```

```python
xr.open_zarr(os.path.join(output_directory, output_filename))
```
