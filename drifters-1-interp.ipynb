This notebook interpolates space-time gridded data at the drifters space-time observation points.

```python
import xarray as xr
```

```python
drifter_ds = xr.open_zarr("data/drifters_drogue15m_CalVal.zarr").compute()
```

```python
duacs_ds = xr.open_zarr("data/cmems_obs-sl_eur_phy-ssh_my_allsat-l4-duacs-0.0625deg_P1D_2023-03-28_2023-07-11.zarr")
```

```python
era5_ds = xr.open_zarr("data/cmems_obs-wind_glo_phy_my_l4_0.125deg_PT1H_2023-03-28_2023-07-11.zarr")
```

```python
drifter_ds["ugos"] = (("points"), duacs_ds.ugos.interp(
    time=drifter_ds.time, latitude=drifter_ds.latitude, longitude=drifter_ds.longitude, method="linear"
).values)
drifter_ds["vgos"] = (("points"), duacs_ds.vgos.interp(
    time=drifter_ds.time, latitude=drifter_ds.latitude, longitude=drifter_ds.longitude, method="linear"
).values)
```

```python
drifter_ds["eastward_stress"] = (("points"), era5_ds.eastward_stress.interp(
     time=drifter_ds.time, latitude=drifter_ds.latitude, longitude=drifter_ds.longitude, method="linear"
).values)
drifter_ds["northward_stress"] = (("points"), era5_ds.northward_stress.interp(
     time=drifter_ds.time, latitude=drifter_ds.latitude, longitude=drifter_ds.longitude, method="linear"
).values)
```

```python
drifter_ds.to_zarr("data/drifters_drogue15m_CalVal.zarr", compute=True, mode="w")
```
